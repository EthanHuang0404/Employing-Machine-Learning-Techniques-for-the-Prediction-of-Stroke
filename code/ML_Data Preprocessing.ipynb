##### 處理類別型資料

```python
import pandas as pd

def impute_with_mode_train(dataframe, id_column='pseudo_no'):
    """
    使用眾數填補訓練集中每一欄的缺失值。

    參數:
        dataframe (pd.DataFrame): 要填補的資料框架
        id_column (str, optional): DataFrame中的ID欄位名稱，預設為'pseudo_no'

    返回:
        pd.DataFrame: 已填補的資料框架
        dict: 特征與其眾數的映射
    """
    # 保留ID欄位
    id_col = dataframe[id_column].copy()

    # 複製資料以避免直接更改原始資料並去除ID欄位
    df_copy = dataframe.drop(columns=[id_column])

    mode_dict = {}
    for column in df_copy.columns:
        mode_value = df_copy[column].mode()[0]
        df_copy[column].fillna(mode_value, inplace=True)
        mode_dict[column] = mode_value

    # 將ID欄位插回到最前面的位置
    df_copy.insert(0, id_column, id_col)

    return df_copy, mode_dict

def impute_with_mode_test(dataframe, mode_dict, id_column='pseudo_no'):
    """
    使用訓練集中計算出的眾數填補測試集。

    參數:
        dataframe (pd.DataFrame): 要填補的資料框架
        mode_dict (dict): 特征與其眾數的映射
        id_column (str, optional): DataFrame中的ID欄位名稱，預設為'pseudo_no'

    返回:
        pd.DataFrame: 已填補的資料框架
    """
    # 保留ID欄位
    id_col = dataframe[id_column].copy()

    # 複製資料以避免直接更改原始資料並去除ID欄位
    df_copy = dataframe.drop(columns=[id_column])

    for column, mode_value in mode_dict.items():
        df_copy[column].fillna(mode_value, inplace=True)

    # 將ID欄位插回到最前面的位置
    df_copy.insert(0, id_column, id_col)

    return df_copy

def one_hot_encode_features(dataframe, features_to_encode, id_column='pseudo_no'):
    """
    對指定的特徵欄位進行one-hot encoding。

    參數:
        dataframe (pd.DataFrame): 要進行one-hot encoding的資料框架
        features_to_encode (list): 要進行one-hot encoding的特徵欄位名稱列表
        id_column (str, optional): DataFrame中的ID欄位名稱，預設為'pseudo_no'

    返回:
        pd.DataFrame: 已進行one-hot encoding的資料框架
    """
    # 保留ID欄位
    id_col = dataframe[id_column].copy()

    # 去除ID欄位
    dataframe = dataframe.drop(columns=[id_column])

    encoded_df = pd.get_dummies(dataframe, columns=features_to_encode, dtype=int)
    encoded_df = encoded_df * 1

    # 將ID欄位插回到最前面的位置
    encoded_df.insert(0, id_column, id_col)

    return encoded_df
```

##### 處理數值型資料

```python
from sklearn.preprocessing import StandardScaler

def fillna_with_mean(dataframe, id_column='pseudo_no'):
    id_col = dataframe[id_column].copy()
    df_copy = dataframe.drop(columns=[id_column])
    fill_values = {}

    for column in df_copy.columns:
        mean_value = df_copy[column].mean()
        df_copy[column].fillna(mean_value, inplace=True)
        fill_values[column] = mean_value

    df_copy.insert(0, id_column, id_col)
    return df_copy, fill_values

def standardize_numeric_features(dataframe, features_to_standardize, id_column='pseudo_no'):
    id_col = dataframe[id_column].copy()
    scaler = StandardScaler()
    standardized_features = scaler.fit_transform(dataframe[features_to_standardize])
    standardized_df = pd.DataFrame(standardized_features, columns=features_to_standardize)
    standardized_df.insert(0, id_column, id_col)
    return standardized_df, scaler
```

```python
# 針對結構化類別型資料做前處理
def process_structured_categorical_data_train(dataframe, features_to_encode):
    """
    處理結構化類別型資料: 進行填補缺失值與one-hot encoding。

    參數:
        dataframe (pd.DataFrame): 要進行處理的資料框架
        features_to_encode (list): 要進行one-hot encoding的特徵欄位名稱列表

    返回:
        pd.DataFrame: 處理後的資料框架
    """
    # 使用眾數填補缺失值
    df_filled, mode_dict = impute_with_mode_train(dataframe)

    # 進行one-hot encoding
    df_encoded = one_hot_encode_features(df_filled, features_to_encode)

    return df_encoded, mode_dict

# 將X_train的標準套用到X_test上
def process_structured_categorical_data_test(dataframe, features_to_encode, mode_dict):
    """
    處理結構化類別型資料: 進行填補缺失值與one-hot encoding。

    參數:
        dataframe (pd.DataFrame): 要進行處理的資料框架
        features_to_encode (list): 要進行one-hot encoding的特徵欄位名稱列表

    返回:
        pd.DataFrame: 處理後的資料框架
    """
    # 使用眾數填補缺失值
    df_filled = impute_with_mode_test(dataframe, mode_dict)

    # 進行one-hot encoding
    df_encoded = one_hot_encode_features(df_filled, features_to_encode)

    return df_encoded


# 針對結構化連續型資料做前處理
def process_structured_continuous_data_train(dataframe, features_to_process, id_column='pseudo_no'):
    df_selected = dataframe[features_to_process + [id_column]].copy()
    df_filled, fill_values = fillna_with_mean(df_selected, id_column)
    df_standardized, scaler = standardize_numeric_features(df_filled, features_to_process, id_column)
    dataframe.loc[:, features_to_process] = df_standardized[features_to_process].values
    return dataframe, fill_values, scaler

# 將X_train的標準套用到X_test上
def process_structured_continuous_data_test(dataframe, fill_values, scaler, features_to_process, id_column='pseudo_no'):
    df_selected = dataframe[features_to_process + [id_column]].copy()

    for feature, value in fill_values.items():
        df_selected[feature].fillna(value, inplace=True)

    id_col = df_selected[id_column].copy()
    standardized_features = scaler.transform(df_selected[features_to_process])
    standardized_df = pd.DataFrame(standardized_features, columns=features_to_process)
    standardized_df.insert(0, id_column, id_col)
    dataframe.loc[:, features_to_process] = standardized_df[features_to_process].values
    return dataframe
```

```python
def align_columns(train_df, test_df):
    """
    確保訓練集和測試集具有相同的列，並按相同的順序排序。

    參數:
        train_df (pd.DataFrame): 訓練集資料框架
        test_df (pd.DataFrame): 測試集資料框架

    返回:
        pd.DataFrame, pd.DataFrame: 調整後的訓練集和測試集資料框架
    """

    # 獲取訓練集和測試集中缺少的列
    missing_cols_train = set(test_df.columns) - set(train_df.columns)
    missing_cols_test = set(train_df.columns) - set(test_df.columns)

    # 在訓練集和測試集中添加缺少的列
    for col in missing_cols_train:
        train_df[col] = 0
    for col in missing_cols_test:
        test_df[col] = 0

    # 確保訓練集和測試集的列順序相同
    train_df = train_df[test_df.columns]

    return train_df, test_df
```

```python
from ckip_transformers.nlp import CkipWordSegmenter, CkipPosTagger, CkipNerChunker

# Initialize drivers
ws_driver  = CkipWordSegmenter(model="bert-base")
pos_driver = CkipPosTagger(model="bert-base")

# Use CPU
#ws_driver = CkipWordSegmenter(device=-1)

# Use GPU:0
ws_driver = CkipWordSegmenter(device=0)
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer, TfidfTransformer
import string
import unicodedata

# 讀取自定義停用詞txt檔
def load_custom_stopwords(custom_stopwords_path):
    with open(custom_stopwords_path, 'r', encoding='utf-8') as f:
        custom_stopwords = f.read().splitlines()
    return custom_stopwords

# 轉換全形到半形字元的函數
def fullwidth_to_halfwidth(text):
    return ''.join([c for c in unicodedata.normalize('NFKC', text)])

# 移除標點符號
def remove_punctuation(text):
    return text.translate(str.maketrans('', '', string.punctuation))

def ckip_segmentation(text_data):
    segmented_text_data = ws_driver(text_data)
    #pos_data = pos_driver(segmented_text_data)
    #return segmented_text_data, pos_data
    return segmented_text_data

def preprocess_text(text_data, custom_stopwords_path):
    custom_stopwords = load_custom_stopwords(custom_stopwords_path)

    # 轉換全形字元和移除標點符號
    text_data = [fullwidth_to_halfwidth(text) for text in text_data]
    text_data = [remove_punctuation(text) for text in text_data]

    segmented_texts = ckip_segmentation(text_data)
    # segmented_texts, pos_texts = ckip_segmentation(text_data)

    # processed_texts = []
    # for idx, segmented_text in enumerate(segmented_texts):
    #     words = segmented_text
    #     pos_tags = pos_texts[idx]
    #     # 只保留特定詞性的字詞
    #     words = [word for word, pos_tag in zip(words, pos_tags) if pos_tag in ['Nc', 'Nv', 'V', 'VA', 'Na']]
    #     # 去除停用詞
    #     filtered_words = [word for word in words if word not in custom_stopwords]
    #     processed_texts.append(' '.join(filtered_words))

    # return processed_texts
    return [" ".join(text) for text in segmented_texts]

def get_bag_of_words_vectorizer(df, custom_stopwords_path, n, m, min_freq, max_freq, fit=True, fitted_vectorizer=None):
    """
    Given a dataframe, return a bag of words dataframe with specified n-gram and frequency threshold.
    If fit is True, the vectorizer will be fitted to the provided data.
    If fit is False, the provided fitted_vectorizer will be used to transform the data.
    """
    processed_texts = preprocess_text(df['triage_cc'], custom_stopwords_path)

    # Use the provided fitted_vectorizer if fit is False
    if fit:
        n_gram = (n, m)
        vectorizer = CountVectorizer(ngram_range=n_gram)
        bag_of_words = vectorizer.fit_transform(processed_texts)
    else:
        if fitted_vectorizer is None:
            raise ValueError("Please provide a fitted vectorizer when fit is set to False.")
        vectorizer = fitted_vectorizer
        bag_of_words = vectorizer.transform(processed_texts)

    bag_of_words_df = pd.DataFrame(bag_of_words.toarray(), columns=vectorizer.get_feature_names_out())
    word_frequencies = bag_of_words_df.mean()
    selected_features = word_frequencies[(word_frequencies >= min_freq) & (word_frequencies <= max_freq)]
    selected_bag_of_words_df = bag_of_words_df[selected_features.index]
    selected_bag_of_words_df['pseudo_no'] = df['pseudo_no'].values

    # Ensure 'pseudo_no' is the first column
    cols = selected_bag_of_words_df.columns.tolist()
    cols = cols[-1:] + cols[:-1]
    selected_bag_of_words_df = selected_bag_of_words_df[cols]

    return selected_bag_of_words_df, vectorizer
```

```python
def convert_to_tfidf_df(bag_of_words_df, fit=True, fitted_transformer=None):
    """
    Convert a bag of words dataframe into a TF-IDF dataframe.
    If fit is True, the transformer will be fitted to the provided data.
    If fit is False, the provided fitted_transformer will be used to transform the data.
    """
    pseudo_no_column = bag_of_words_df['pseudo_no']
    bag_of_words_df = bag_of_words_df.drop('pseudo_no', axis=1)
    term_frequency_matrix = bag_of_words_df.values

    if fit:
        tfidf_transformer = TfidfTransformer()
        tfidf_matrix = tfidf_transformer.fit_transform(term_frequency_matrix)
    else:
        if fitted_transformer is None:
            raise ValueError("Please provide a fitted transformer when fit is set to False.")
        tfidf_transformer = fitted_transformer
        tfidf_matrix = tfidf_transformer.transform(term_frequency_matrix)

    tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=bag_of_words_df.columns)
    tfidf_df.insert(0, 'pseudo_no', pseudo_no_column)

    return tfidf_df, tfidf_transformer
```

##### Bag of Words

1. 計算每個詞彙在所有病歷中出現的次數。
2. 計算每個詞彙在所有病歷中出現的頻率（出現次數除以總病歷數）。
3. 根據95%和5%的閾值將詞彙選取或剔除。
4. 將選取的詞彙轉換回DataFrame以供後續操作。

```python
import pandas as pd
import spacy
import re
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer, TfidfTransformer

# 載入英文模型
nlp = spacy.load("en_core_web_sm")

def lemmatize_text(text):
    doc = nlp(text)
    lemmatized_text = " ".join([token.lemma_ for token in doc if not token.is_stop and not token.like_num])
    return lemmatized_text

def get_bag_of_words_vectorizer_en(df, custom_stopwords_path, n, m, min_freq, max_freq, fit=True, fitted_vectorizer=None):
    """
    Given a dataframe, return a bag of words dataframe with specified n-gram and frequency threshold.
    If fit is True, the vectorizer will be fitted to the provided data.
    If fit is False, the provided fitted_vectorizer will be used to transform the data.
    """
    # 讀取自訂義停用詞txt檔
    with open(custom_stopwords_path, 'r') as f:
        custom_stopwords = f.read().splitlines()

    # 從df中提取PI
    text_data = df['PI'].tolist()

    # 進行日期時間去除、斷詞和詞形還原
    processed_texts = []
    for text in text_data:
        # 使用正規表達式刪除日期時間
        text_without_datetime = re.sub(r'\d{4}/\d{2}/\d{2} \d{2}:\d{2}:\d{2}', '', text)
        # 進行詞形還原
        lemmatized_text = lemmatize_text(text_without_datetime)
        processed_texts.append(lemmatized_text)

    # Use the provided fitted_vectorizer if fit is False
    if fit:
        n_gram = (n, m)
        vectorizer = CountVectorizer(ngram_range=n_gram, stop_words=custom_stopwords)
        bag_of_words = vectorizer.fit_transform(processed_texts)
    else:
        if fitted_vectorizer is None:
            raise ValueError("Please provide a fitted vectorizer when fit is set to False.")
        vectorizer = fitted_vectorizer
        bag_of_words = vectorizer.transform(processed_texts)

    bag_of_words_df = pd.DataFrame(bag_of_words.toarray(), columns=vectorizer.get_feature_names_out())
    word_frequencies = bag_of_words_df.mean()
    selected_features = word_frequencies[(word_frequencies >= min_freq) & (word_frequencies <= max_freq)]
    selected_bag_of_words_df = bag_of_words_df[selected_features.index]
    selected_bag_of_words_df['pseudo_no'] = df['pseudo_no'].values

    # Ensure 'pseudo_no' is the first column
    cols = selected_bag_of_words_df.columns.tolist()
    cols = cols[-1:] + cols[:-1]
    selected_bag_of_words_df = selected_bag_of_words_df[cols]

    return selected_bag_of_words_df, vectorizer
```

```python
def convert_to_tfidf_df(bag_of_words_df):
    # 提取 'pseudo_no' column 並稍後使用
    pseudo_no_col = bag_of_words_df['pseudo_no'].copy()

    # 從bag_of_words_df中刪除 'pseudo_no' column
    bag_of_words_df = bag_of_words_df.drop('pseudo_no', axis=1)

    # 提取詞頻矩陣
    term_frequency_matrix = bag_of_words_df.values

    # 初始化TfidfTransformer並計算TF-IDF權重
    tfidf_transformer = TfidfTransformer()
    tfidf_matrix = tfidf_transformer.fit_transform(term_frequency_matrix)

    # 將TF-IDF權重轉換為DataFrame，並加上詞彙作為欄位名稱
    tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=bag_of_words_df.columns)

    # 將pseudo_no加到DataFrame的最前面
    tfidf_df.insert(0, 'pseudo_no', pseudo_no_col.values)

    return tfidf_df
```

```python
from sklearn.feature_extraction.text import CountVectorizer

def get_count_vectorized_data(train_data, test_data):
    train_data = train_data.fillna('0')
    test_data = test_data.fillna('0')

    # Initialize and fit the CountVectorizer
    vectorizer = CountVectorizer(tokenizer=lambda x: x.split(' '))
    train_transformed = vectorizer.fit_transform(train_data)

    # Transform the test data using the same vectorizer
    test_transformed = vectorizer.transform(test_data)

    # Convert to dataframes
    train_df = pd.DataFrame(train_transformed.toarray(), columns=vectorizer.get_feature_names_out())
    test_df = pd.DataFrame(test_transformed.toarray(), columns=vectorizer.get_feature_names_out())

    # Ensure test_df has the same columns as train_df
    missing_cols = set(train_df.columns) - set(test_df.columns)
    for c in missing_cols:
        test_df[c] = 0
    test_df = test_df[train_df.columns]

    return train_df, test_df
```

```python
def concat_dataframes_on_pseudo_no(dfs, categorical_dfs):
    """
    Concatenate multiple dataframes on the 'pseudo_no' column and ensure
    columns from specified dataframes are of integer type.

    Parameters:
        dfs (list of pd.DataFrame): List of dataframes to concatenate.
        categorical_dfs (list of pd.DataFrame): List of dataframes that contain categorical columns.

    Returns:
        pd.DataFrame: Concatenated dataframe.
    """
    categorical_columns = [col for df in categorical_dfs for col in df.columns if col != 'pseudo_no']

    merged_df = dfs[0]
    for df in dfs[1:]:
        merged_df = pd.merge(merged_df, df, on='pseudo_no', how='outer')

    # Ensure columns from categorical_dfs are of integer type
    for col in categorical_columns:
        if col in merged_df.columns and merged_df[col].dtype == float:
            merged_df[col] = merged_df[col].fillna(0).astype(int)

    return merged_df
```
